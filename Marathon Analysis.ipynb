# Exploratory Data Analysis on a Marathon Dataset

```python
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
marathon_data = pd.read_csv("Marathon.csv")
```

### Clean up the data

```python
Filt_marathon_data = marathon_data[(marathon_data['Event distance/length'].isin(['100km', '100mi'])) & (marathon_data['Year of event'] == 2018)] 
Filt_marathon_data # Combined and filtered data for only 100km and 100mi races for year 2018
```

```python
Filt_marathon_data_GER = Filt_marathon_data[Filt_marathon_data['Event name'].str.endswith('(GER)', na = False)]
Filt_marathon_data_GER # Filtered for only where the country is Germany
```

```python
Filt_marathon_data_GER.loc[:, 'Event name'] = Filt_marathon_data_GER['Event name'].str.replace(r'\(.*\)', '' , regex = True)
Filt_marathon_data_GER #The trailing (GER) was removed from the Event name column
```

```python
#Create a new column 'athlete_age'
Filt_marathon_data_GER.loc[:,'Athlete_age'] = Filt_marathon_data_GER['Year of event'] - Filt_marathon_data_GER['Athlete year of birth']
Filt_marathon_data_GER.head(5)
```

```python
#Remove h from athlete performance
Filt_marathon_data_GER.loc[:,'Athlete performance'] = Filt_marathon_data_GER['Athlete performance'].str.rstrip('h')
#Filt_marathon_data_GER.loc[:, 'Athlete performance'] = Filt_marathon_data_GER.loc['Athlete performance'].str.split(' ').str.get(0)
Filt_marathon_data_GER.head(5)
```

```python
#Delete columns: Athlete Club, Athlete Country, Athlete year of birth, Athlete Age Category
Filt_marathon_data_GER = Filt_marathon_data_GER.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis = 1)
Filt_marathon_data_GER.head(5)
```

```python
#Check for null values
Filt_marathon_data_GER.isna().sum()
```

```python
Filt_marathon_data_GER[Filt_marathon_data_GER['Athlete_age'].isna()==True]
```

```python
#Delete the NaN rows
Filt_marathon_data_GER = Filt_marathon_data_GER.dropna()
```

```python
#Check for duplicates
Filt_marathon_data_GER[Filt_marathon_data_GER.duplicated() == True]
```

```python
#Reset Index
Filt_marathon_data_GER.reset_index(drop = True) # creates new index and deletes the old one
```

```python
#Change types
Filt_marathon_data_GER.dtypes
```

```python
Filt_marathon_data_GER['Athlete_age'] = Filt_marathon_data_GER['Athlete_age'].astype(int)
```

```python
Filt_marathon_data_GER['Athlete average speed'] = Filt_marathon_data_GER['Athlete average speed'].astype(float)
```

```python
Filt_marathon_data_GER.dtypes
```

```python
Filt_marathon_data_GER.head(5)
```

```python
#Rename columns
Filt_marathon_data_GER = Filt_marathon_data_GER.rename(columns = {'Year of event': 'Year',
                                                                  'Event dates': 'MarathonDay',
                                                                  'Event name':  'MarathonName',
                                                                  'Event distance/length': 'MarathonLength',
                                                                  'Event number of finishers': 'MarathonFinishers',
                                                                  'Athlete performance' : 'AthletePerformance',
                                                                  'Athlete gender' : 'AthleteGender',
                                                                  'Athlete average speed' : 'AthleteAverageSpeed',
                                                                  'Athlete ID' : 'AthleteID',   
                                                                  'Athlete_age' : 'AthleteAge' 
})
```

```python
Filt_marathon_data_GER.head(5)
```

```python
#Reorder the columns
Marathon_GER = Filt_marathon_data_GER[['MarathonDay', 'MarathonName', 'MarathonLength', 'MarathonFinishers', 'AthleteID', 'AthleteGender', 'AthleteAge', 'AthletePerformance', 'AthleteAverageSpeed', 'Year']]
Marathon_GER.head(5)
```

### Charts and Graphs


```python
#Replace infinite values with NaN
Marathon_GER['MarathonLength'] = Marathon_GER['MarathonLength'].replace([float('inf'), float('-inf')], pd.NA)
#Plot the histogram
sns.histplot(Marathon_GER['MarathonLength'].dropna())
```

```python
#ensuring that only rows with NaN values in MarathonLength and AthleteGender are removed before plotting
sns.histplot(Marathon_GER.dropna(subset=['MarathonLength', 'AthleteGender']), x='MarathonLength', hue='AthleteGender')
```

```python
#distribution of 100mil by athlete average speed
sns.displot(Marathon_GER[Marathon_GER['MarathonLength'] == '100mi']['AthleteAverageSpeed'])
```

```python
#Plot of speed against the various athlete ages
sns.lmplot(data = Marathon_GER, x = 'AthleteAge', y='AthleteAverageSpeed', hue = 'AthleteGender')
```

```python
#Difference in speed for the 100km 100mi between male to female
Marathon_GER.groupby(['MarathonLength', 'AthleteGender'])['AthleteAverageSpeed'].mean()
```
